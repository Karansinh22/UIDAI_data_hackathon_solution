# 01. Geographic Disparity - Analysis

This notebook performs geographic analysis on Aadhaar enrollment and updates to find underserved regions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Setting plot style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

os.makedirs('../../visualizations', exist_ok=True)
print("Libraries imported and visualization directory ready.")
```

## 1. Load Processed Data

```python
data = pd.read_csv('../../processed_data/geographic_data.csv')
print(f"Loaded {len(data)} records for analysis.")
```

## 2. Top 10 States by Enrollment

```python
state_data = data.groupby('state')['total_enrollments'].sum().sort_values(ascending=False).head(10)
state_data.plot(kind='bar', color='skyblue')
plt.title('Top 10 States by Total Enrollments')
plt.ylabel('Enrollments')
plt.savefig('../../visualizations/01_top_10_states_enrollment.png')
plt.show()
```

## 3. District Performance Heatmap (Sample)
We'll look at the distribution of enrollments across districts in a major state.

```python
sample_state = data['state'].iloc[0]
state_districts = data[data['state'] == sample_state].groupby('district')['total_enrollments'].sum().sort_values(ascending=False).head(20)

sns.barplot(x=state_districts.values, y=state_districts.index, palette='viridis')
plt.title(f'Enrollment Distribution in {sample_state} (Top 20 Districts)')
plt.xlabel('Enrollments')
plt.savefig('../../visualizations/01_district_distribution_sample.png')
plt.show()
```

## 4. Updates vs Enrollments Correlation

```python
sns.scatterplot(data=data, x='total_enrollments', y='total_updates', alpha=0.5)
plt.title('Enrollments vs Total Updates by Pincode')
plt.savefig('../../visualizations/01_enrollments_vs_updates_correlation.png')
plt.show()
```
